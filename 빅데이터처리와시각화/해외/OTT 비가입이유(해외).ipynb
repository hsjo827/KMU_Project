```python
# 데이터 불러오기
df = pd.read_csv('방송통신위원회_해외방송시장조사_유료 OTT 서비스 비가입 이유_20220711.csv', encoding='euc-kr')
df.head()
```

```python
# 데이터프레임 df에서 '구분'이 '전체'인 행을 선택하고 '국가' 열을 인덱스로 설정
tem = df.query("구분=='전체'").set_index('국가').T.iloc[5:,:]
tem = tem.T

# 새로운 열 '기타'를 생성하여, 특정 열들의 합계를 넣음
tem['기타']=tem['이용방법을잘몰라서(비율)']+tem['콘텐츠품질이떨어져서(비율)']+tem['온라인동영상서비스(OTT)가만족스럽지않아서(비율)']+tem['자녀들에게부정적인영향을미칠것같아서(비율)']
# 합계에 포함된 열들을 데이터프레임에서 삭제
tem = tem.drop(['이용방법을잘몰라서(비율)', '콘텐츠품질이떨어져서(비율)','온라인동영상서비스(OTT)가만족스럽지않아서(비율)','자녀들에게부정적인영향을미칠것같아서(비율)'], axis=1)

tem
```
